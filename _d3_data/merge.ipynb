```python
d3_5countries = d3[d3['id'].isin(['USA', 'BRA', 'IDN', 'IND', 'IRN'])]
```

```python
d3_5countries.head(3)
```

```python
d3_5countries['confirmed_merge_ratio'] = (d3_5countries['Cases - newly reported in last 24 hours'] * 7 - d3_5countries['Cases - newly reported in last 7 days']) / d3_5countries['Cases - newly reported in last 7 days']
d3_5countries['death_merge_ratio'] = (d3_5countries['Deaths - newly reported in last 24 hours'] * 7 - d3_5countries['Deaths - newly reported in last 7 days']) / d3_5countries['Deaths - newly reported in last 7 days']
```

```python
d3_5countries[['Name', 'death_merge_ratio']]
```

```python
d3_5countries[['Name', 'confirmed_merge_ratio']]
```

```python
d1 = d3_5countries[['Name', 'death_merge_ratio']].rename(columns={'death_merge_ratio' : 'ratio'})
d2 = d3_5countries[['Name', 'confirmed_merge_ratio']].rename(columns={'confirmed_merge_ratio' : 'ratio'})
tot = pd.concat([d1, d2], axis=0)
```

```python
tot = tot.sort_values(by='Name').reset_index()
```
